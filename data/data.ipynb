```python
def get_ordered_columns(level=10, direct=("Bid","Ask"), pv=("Price", "Volume"),index=True,price=True,volume=True):
    """
    Get the ordered columns for the level of the order book.
    """

    columns = []

    if index:
        columns.append("index")

    if price:
        for i in range(level,0,-1):
            columns.append(f"{direct[0]}{pv[0]}{i}")
        for i in range(1,level+1):
            columns.append(f"{direct[1]}{pv[0]}{i}")

    if volume:
        for i in range(level,0,-1):
            columns.append(f"{direct[0]}{pv[1]}{i}")
        for i in range(1,level+1):
            columns.append(f"{direct[1]}{pv[1]}{i}")

    return columns


price_columns = get_ordered_columns(index=False, price=True, volume=False)
volume_columns = get_ordered_columns(index=False, price=False, volume=True)
print(price_columns)
print(volume_columns)
```

```python

price_data = a.data[price_columns].values
print(price_data.shape)
# 去掉所有nan值后的平均值
print(np.nanmean(price_data))
print(np.nanstd(price_data))

a.data[price_columns] = (a.data[price_columns] - np.nanmean(price_data))/np.nanstd(price_data)

print(a.data.head(10))

# data/dataset/real_data/processed/sz000001-level10_resampled_zscore_processed.csv
```

```python
import numpy as np
from numpy.lib.stride_tricks import sliding_window_view

def unbalance_data(data, window=100, step=2, dtype=np.float16):
    """
    进行时序数据采样，支持指定窗口大小和步长。
    
    参数：
        data (ndarray): 归一化后的时序数据，形状为 (num_samples, num_features)
        window (int): 采样窗口大小
        step (int): 采样步长
        dtype (numpy dtype): 结果数据类型
    
    返回：
        unbalanced_X (ndarray): 形状为 (num_windows, window, num_features) 的采样数据
    """
    if len(data) < window:
        raise ValueError("数据长度小于窗口大小，无法进行采样。")
    
    # 使用 sliding_window_view 创建滑动窗口
    windows = sliding_window_view(data, (window, data.shape[1]))[::step, 0]
    
    return windows.astype(dtype)

# 示例用法
data = np.array([i for i in range(1000)]).reshape(100,10) 
print(data)# 假设原始数据有 1000 行，5 个特征
sampled_data = unbalance_data(data, window=10, step=2)

print(sampled_data.shape)

print(sampled_data[0])
print(sampled_data[1])# 期望形状: (num_samples, 100, num_features)

d1=data[:10,:]
d2=sampled_data[0]

print(sum(d1==d2))
```
